```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter
import matplotlib.ticker as ticker
import seaborn as sns
import numpy as np
df = pd.read_csv('diabetes.csv')
```

```python
np.random.seed(42)

RandEntriesNumber = 25

RandIndices = np.random.choice(df.index, size=RandEntriesNumber, replace=False)

RandEntries = df.loc[RandIndices]

print('Randomly selected values')
print(f"mean: {RandEntries['Glucose'].mean()}")
print(f"max:  {RandEntries['Glucose'].max()}")
print('')
print('Population values')
print(f"mean: {df['Glucose'].mean()}")
print(f"max:  {df['Glucose'].max()}")

plt.boxplot([RandEntries['Glucose'],df['Glucose']])
plt.title('Randomly Selected values & Population values vs Glucose')
plt.xticks([1,2], ['Randomly Selected Subset','Population'])
plt.show()
```

```python
print(f"98th Percentile for Randomly Selected Subset: {np.percentile(RandEntries['BMI'], 98)}")
print(f"98th Percentile for Randomly Selected Subset: {np.percentile(df['BMI'], 98)}")


plt.boxplot([RandEntries['BMI'],df['BMI']])
plt.ylim(30,70)
plt.title('98th Percentile of Randomly Selected values & Population values vs. BMI')
plt.xticks([1,2], ['Randomly Selected Subset','Population'])
plt.axhline(40.248, color='b', linestyle='--', label='Randomly Selected Values')
plt.axhline(47.526, color='r', linestyle='--', label='Population Values')
plt.legend()
plt.show()
```

```python
np.random.seed(42)
EmptyList = []
for i in range(500):
    RandSamp = df['BloodPressure'].sample(n=150, replace=True)
    EmptyList.append(RandSamp)

RandomSamples = pd.concat(EmptyList)
print('Stats for Randomly Selected Samples')
print(f"Mean: {RandomSamples.mean()}")
print(f"Standard Deviation: {RandomSamples.std()}")
print(f"98th Percentile: {np.percentile(RandomSamples, 98)}")
print('')
print('Stats for the Population')
print(f"Mean: {df['BloodPressure'].mean()}")
print(f"Standard Deviation: {df['BloodPressure'].std()}")
print(f"98th Percentile: {np.percentile(df['BloodPressure'], 98)}")

plt.title('Randomly Selected values & Population values vs. Blood Pressure Stats Comparison')
plt.boxplot([RandomSamples,df['BloodPressure']])
plt.xticks([1,2], ['Randomly Selected Subset','Population'])
plt.axhline(100, color='b', linestyle='--', label='98th Percentile of Randomly Selected Values')
plt.axhline(99.32, color='r', linestyle='--', label='98th percentile of Population Values')
plt.legend()
plt.show()
```
